```python
import requests as rq
import pandas as pd
import re
from bs4 import BeautifulSoup 
```

```python
url = "https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
Bresp = rq.get(url)
```

```python
bSoup = BeautifulSoup(Bresp.content,"html.parser")
```

```python
bSoup
```

```python
containers = bSoup.findAll('div',{'class':'tUxRFH'})
```

```python
print(type(containers), len(containers))
```

```python
column_names = ['ProductName', 'Stars', 'Ratings', 'Reviews', 'CurrentPrice', 'MRP', 'Processor', 'Ram', 'Oprating_System', 'Storage', 'Disply','Warranty']

Laptop_information = []


for container in containers:
    # Finding Product Name
    product = container.find('div',{'class':'KzDlHZ'})
    ProductName = product.text.split('-')[0].strip()

    # Finding Stars
    star = container.find('div',{'class':'XQDdHH'})
    try:
        Stars = star.text
    except:
        Stars = 0
    
    # Finding Ratings & Reviews
    Rating = container.find('span',{'class':'Wphh3N'})
    try:
        ratRev = re.findall(r'\d+,?\d*', Rating.text)
        Ratings = ratRev[0]
        Reviews = ratRev[1]
    
    except:
        Ratings = 0
        Reviews = 0

    
    

    # Finding Current Price
    CurrentPrice = container.find('div', {'class': 'Nx9bqj _4b5DiR'}).text
    #print(CurrentPrice.text)

    # Finding MRP
    mrp = container.find('div',{'class':'yRaY8j ZYYwLA'})
    try:
        MRP = mrp.text
    except:
        MRP = 0

    # Finding information about the product
    info = container.findAll('li', {'class':'J+igdf'})
    Processor = info[0].text
    Ram = info[1].text
    Oprating_System = info[2].text
    Storage = info[3].text
    Disply = info[4].text
    Warranty = info[5].text

    #print(ProductName, Stars, Ratings, Reviews, CurrentPrice, MRP, Processor, Ram, Oprating_System, Storage, Disply,Warranty)
    #print('\n')

    Laptop_information.append([ProductName, Stars, Ratings, Reviews, CurrentPrice, MRP, Processor, Ram, Oprating_System, Storage, Disply,Warranty])
```

```python
df = pd.DataFrame(Laptop_information,columns=column_names)
```
