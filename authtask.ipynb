```python
import requests as rq
```

```python
nUrl = 'https://www.timeanddate.com/'
```

```python
nParms = {
    'lat': 33.2778,
    'lng': 75.3412,
    'timezone' : 'IST',
    'date' : 'toady'
}
```

```python
nHaders = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
}
```

```python
nResp = rq.get(nUrl, params=nParms, headers=nHaders)
```

```python
print(nResp)
```

```python
print(nResp.content)
```
